# Logistic Regression

```python
from sklearn.model_selection import train_test_split
```

```python
X = ad_data[['Daily Time Spent on Site', 'Age', 'Area Income','Daily Internet Usage', 'Male']]
y = ad_data['Clicked on Ad']
```

```python
XTrain, XTest, yTrain, yTest = train_test_split(X, y , test_size = 0.3, random_state = 101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
glm = LogisticRegression()
```

```python
glm.fit(X = XTrain, y = yTrain)
```

## Predictions and Evaluations
** Now predict values for the testing data.**

```python
predictions = glm.predict(X = XTest)
```

** Create a classification report for the model.**

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
cm = pd.DataFrame(confusion_matrix(y_true = yTest, y_pred = predictions), 
                  index = ['0', '1'], columns = ['0', '1'])
cm
```

```python
print(classification_report(y_true = yTest, y_pred = predictions))
```

```python
model_interpret = pd.DataFrame(data = glm.coef_[0], index = X.columns, columns = ['Coefficients'])
model_interpret
```
